## Problem 1 [5+5+5]
### (a)
$\det Q = \pm 1$.  
Since we have  
$$\det Q^{-1} = \det Q^T = \det Q$$
therefore  
$$1 = \det I = \det (QQ^{-1}) = (\det Q)^2 \Rightarrow \det Q = \pm 1$$

### (b)
If $\lambda \in \mathbb R$, then $\lambda = \pm 1$.  
Assume for some real $\lambda$, $x$ is the correspond eigenvector.  
We have both
$$Qx = \lambda x$$
$$x^TQ^T = \lambda x^T$$
Therefore, 
$$x^TQ^TQx = \lambda^2 x^Tx$$
$$\|x\| = \lambda^2 \|x\|$$
Since $x$ is non-zero, $\|x\| > 0$ so $\lambda^2 = 1 \Rightarrow \lambda = \pm 1$.

### (c)
$|\lambda| = 1$.  

```julia
using LinearAlgebra

# generate a random (Haar-uniform) n x n real unitary matrix
# using the algorithm from https://arxiv.org/abs/math-ph/0609050
function randQ(n)
    QR = qr(randn(n,n))
    return QR.Q * Diagonal(sign.(diag(QR.R)))
end
# code outline for 1c:

using Plots

plt = plot!(
    1,
    xlim = (-1.5, 1.5),
    ylim = (-1.5, 1.5),
    title = "problem 1c",
    xlabel = "real part",
    ylabel = "imaginary part",
    aspect_ratio = 1,
    seriestype=:scatter
)

# build an animated gif by pushing new points to the plot, saving every 10th frame

@gif for i = 1:100
    for λ in eigvals(randQ(3))
        push!(plt, real(λ) , imag(λ))
    end
end
```
